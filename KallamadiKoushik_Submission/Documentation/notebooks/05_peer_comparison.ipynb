# B100 Stock Market Listed Companies
## Notebook 5: Peer Similarity Analysis

This notebook calculates similarity scores between all companies using Cosine Similarity on key financial indicators. This allows identifying the closest peer groups for any specific company.

```python
import os
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

clean_path = r"../data/clean"
scores = pd.read_csv(os.path.join(clean_path, "scores_clean.csv"))
companies = scores['symbol'].tolist()

features = ['sales', 'net_profit', 'opm_percentage', 'debt_to_equity', 'roce_percentage', 'roe_percentage']
X = scores[features].fillna(scores[features].median())

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

similarity = cosine_similarity(X_scaled)
sim_df = pd.DataFrame(similarity, index=companies, columns=companies)

peer_mapping = []
for company in companies:
    closest_peers = sim_df[company].sort_values(ascending=False).iloc[1:4].index.tolist()
    peer_mapping.append({
        "symbol": company,
        "peer_1": closest_peers[0],
        "peer_2": closest_peers[1],
        "peer_3": closest_peers[2]
    })

peer_df = pd.DataFrame(peer_mapping)
peer_df.to_csv(r"../data/clean/peer_mapping.csv", index=False)
print("Peer mappings completed. Sample:")
print(peer_df.head(10))
```
