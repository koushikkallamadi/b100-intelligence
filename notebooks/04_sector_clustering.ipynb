# B100 Stock Market Listed Companies
## Notebook 4: Sector Clustering

This notebook segments the Nifty 100 companies into 5 distinct groups using the K-Means clustering algorithm on standard financial metrics.

```python
import os
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

clean_path = r"../data/clean"
scores = pd.read_csv(os.path.join(clean_path, "scores_clean.csv"))

features = ['sales', 'net_profit', 'opm_percentage', 'debt_to_equity', 'roce_percentage', 'roe_percentage']
X = scores[features].fillna(scores[features].median())

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Run K-Means with K=5
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
scores['cluster'] = kmeans.fit_predict(X_scaled)

# Reduce to 2D using PCA for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)
scores['pca1'] = pca_result[:, 0]
scores['pca2'] = pca_result[:, 1]

plt.figure(figsize=(10, 7))
sns.scatterplot(data=scores, x='pca1', y='pca2', hue='cluster', palette='Set1', s=100)
plt.title("Company Financial Clusters (K-Means, K=5) - PCA Visualization")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.show()
```

```python
print("Cluster Profiles (Median values):")
print(scores.groupby('cluster')[features].median())
```
